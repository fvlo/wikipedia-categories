# CATEGORIZATION RULES

## Wikipedia path final category

```python
# Wikipedia category level 1 mapping
# See spreadsheet https://docs.google.com/spreadsheets/d/1c4g4MujTgaqTn5wG5I-ARY06g7rlnHst0whojsFbps8/edit?usp=sharing
MAPPING_wiki_to_trivia = {'History':'History and society' ,'Reference':'None' ,'Law':'History and society' ,'Religion':'History and society' ,'Education':'History and society' ,'Philosophy':'Science and nature' ,'Politics':'History and society' ,'Science_and_technology':'Science and nature' ,'Sports':'Sports and leisure' ,'Language':'Art and literature' ,'Nature':'Science and nature' ,'Geography':'Geography and places' ,'Arts':'Art and literature' ,'Society':'History and society' ,'Culture':'History and society' ,'Health':'Science and nature' ,'Mathematics':'Science and nature' ,'Universe':'Science and nature' ,'Events':'History and society' ,'Humanities':'Art and literature' ,'Life':'Science and nature' ,'Entertainment':'Entertainment' }
```

```python
t_dataRun["WikiPath_Category"] = t_dataRun["WikiCat_1"].replace(MAPPING_wiki_to_trivia)
```

## Wikipedia path exceptions

```python
# https://docs.google.com/spreadsheets/d/1c4g4MujTgaqTn5wG5I-ARY06g7rlnHst0whojsFbps8/edit#gid=2055637170
# Mapping list
exceptionCategories = [('3', 'Literature', 'Art and literature'),('2', 'Food_and_drink', 'Sports and leisure'),('5', 'Plants', 'Science and nature'),('3', 'Ethnobiology', 'Science and nature'),('2', 'Artificial_objects', 'Science and nature'),('4', 'Diseases_and_disorders', 'Science and nature'),('2', 'Performing_arts', 'Art and literature'),('2', 'Gaming', 'Entertainment'),('4', 'Anthropology_of_religion', 'History and society'),('3', 'Leisure', 'Sports and leisure')]
```

```python
# Take dataframe row as input
# If several exceptions apply, algorithm returns the one listed last in exceptionCategories
def findExceptionWikiCatRows(a):
    
    result = ""
    
    for exep in exceptionCategories:
        if a["WikiCat_" + str(exep[0])] == exep[1]:
            result = exep[2]
    
    return result
    
```

```python
row = t_dataRun.iloc[4277]
findExceptionWikiCatRows(row)
```

```python
%%time
t_dataRun["WikiPath_Exceptions"] = t_dataRun.apply(lambda x: findExceptionWikiCatRows(x), axis = 1)
```

```python
%%time
# If exception exists, use that, else use WikiPath_Category
t_dataRun["WikiPath_withExceptions"] = t_dataRun.apply(lambda x: x["WikiPath_Exceptions"] if x["WikiPath_Exceptions"] != "" else x["WikiPath_Category"], axis = 1)
```

```python
correctShares = t_dataRun.loc[:, ["CONS_category", "WikiPath_Category"]]
```

```python
correctShares = correctShares.where(correctShares.CONS_category != "Uncategorized").dropna()
```

```python
correctShares["isSame"] = correctShares.CONS_category == correctShares.WikiPath_Category
```

```python
correctShares = correctShares.groupby([correctShares.WikiPath_Category, correctShares.isSame]).count()
```

```python
correctShares = correctShares.groupby(level = 0).apply(lambda x: 100 * x / float(x.sum()))
```

```python
correctShares = correctShares.unstack(level = -1)
```

```python
correctShares.columns = correctShares.columns.get_level_values(1)
```

```python
correctShares.iloc[:,::-1].plot.barh(stacked = True)
```
